```python
# Guarda todas las IDs de pedidos que tengan un valor ausente en 'add_to_cart_order'
missing_cart_ids = order_products_df[order_products_df['add_to_cart_order'].isnull()]['order_id'].unique()
print("Total de pedidos con 'add_to_cart_order' ausente:", len(missing_cart_ids))
```

```python
# ¿Todos los pedidos con valores ausentes tienen más de 64 productos?
# Agrupa todos los pedidos con datos ausentes por su ID de pedido.
# Cuenta el número de 'product_id' en cada pedido y revisa el valor mínimo del conteo.
# Filtrar los pedidos ausentes
missing_orders = order_products_df[order_products_df['order_id'].isin(missing_cart_ids)]

# Contar productos por pedido
product_counts = missing_orders.groupby('order_id')['product_id'].count()

# Verificar si todos tienen más de 64 productos
all_above_64 = (product_counts > 64).all()
print("¿Todos los pedidos con datos ausentes tienen más de 64 productos?", all_above_64)
# Conteo mínimo de productos por pedido entre los pedidos con valores ausentes
min_count = product_counts.min()
print("Conteo mínimo de productos por pedido (con datos ausentes):", min_count)
```

### [C3] Para cada producto, ¿cuál es la proporción de las veces que se pide y que se vuelve a pedir?

```python
# Total y repetidos por producto
product_orders = order_products_df.groupby('product_id')['reordered'].agg(['sum', 'count'])
product_orders['reorder_rate'] = product_orders['sum'] / product_orders['count']

# Mostrar top 10 con al menos 100 órdenes para evitar sesgo por bajo volumen
filtered = product_orders[product_orders['count'] >= 100].sort_values('reorder_rate', ascending=False)
top_product_reorder_rate = filtered.head(10).merge(products_df, on='product_id')[['product_id', 'product_name', 'reorder_rate']]
print(top_product_reorder_rate)
```

Escribe aquí tus conclusiones Los productos con las mayores proporciones de reorden, es decir, la tasa a la que se vuelven a pedir tras la primera compra, incluyen principalmente productos lácteos y bebidas saludables. Otros productos como diferentes tipos de leche orgánica y pan integral también muestran altas tasas de reorden, todas superiores al 85%, lo que indica una fuerte preferencia y fidelidad hacia estos productos.
